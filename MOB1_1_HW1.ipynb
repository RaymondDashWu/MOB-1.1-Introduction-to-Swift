1. Write a function that returns the sum of all cells but exclude all odd rows

image.png

```
let grid = [[0,0,0,0,0],
            [0,1,2,3,4],
            [0,2,4,6,8],
            [0,3,6,9,12],
            [0,4,8,12,16]]
```

```

func calculateEveryOtherGridTotal(grid: [[Int]]) -> Int {
  var total = 0
  for ind in 0...grid.count - 1 {
    if ind % 2 == 0 {
      for rowItem in grid[ind] {
        // print(ind,rowItem)
        total += rowItem
      }
    }
  }

  return total
}

print(calculateEveryOtherGridTotal(grid: grid))
```

3. Write a program that prints the powers of 2 that are less than or equal to N


```
import Foundation

func twoPowerLessThan(N: Double) -> [Double] {
  var resultArray = [Double]()
  var counter = 0.0
  while pow(2, counter) <= N {
    resultArray.append(pow(2, counter))
    counter += 1
  }
  return resultArray
}

print(twoPowerLessThan(N: 100))
```
